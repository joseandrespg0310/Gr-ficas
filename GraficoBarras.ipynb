# Diagrama de Dispersion

```python
# Elabora un Diagrama de Dispersion
import numpy as np
plt.plot(np.arange(0,10),np.arange(0,10)**3,'bo'#Marca Circulo Azul
         )
# xmin, ymin, ymax
plt.axis([0,10,0,1000])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# CorrecciÃ³n de arange
a = np.arange(-10, 10, 0.3)

# CorrecciÃ³n de plt.plot
plt.plot(a, a**2, 'y^',  # Marca Triangular
         label='a^2')
plt.plot(a, a**3, 'bo', label='a^3')  # Marca circular
plt.plot(a, a**4, 'r--', label='a^4')  # LÃ­nea punteada roja

# Ajuste de rangos
plt.axis([-10, 10, -10, 70])

plt.legend()
plt.show()
```

```python
# Grafico dentro de otro grafico
a=np.arange(1,10,1)
# Primer grafico: 1 fila y 2 columnas
# Para crear un subgrafico se usa subplot
plt.subplot(1,2,1)
plt.plot([1,2,3,4,5],[1,8,27,64,125],'y^') # Con esto se hace la imagen
#Segundo Grafico 1 fila y columnas
plt.subplot(1,2,2)
plt.plot(a,a**3,'bo',a,a**4,'r--')

```

# Histograma

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos: distribuciÃ³n normal estÃ¡ndar con 1000 nÃºmeros aleatorios
datos = np.random.normal(0, 1, 1000)

# Crear el histograma con un color vÃ¡lido
plt.hist(datos, color='#7F38CC', bins=10, edgecolor='black')

# Mejoras en el grÃ¡fico
plt.title("Histograma de una DistribuciÃ³n Normal EstÃ¡ndar")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# AÃ±adir texto explicativo en el grÃ¡fico
media = np.mean(datos)
desviacion = np.std(datos)
plt.text(-3, 80, f"Media: {media:.2f}\nDesviaciÃ³n estÃ¡ndar: {desviacion:.2f}", 
         fontsize=10, color='black', bbox=dict(facecolor='white', alpha=0.5))

# Mostrar el grÃ¡fico
plt.show()
```
